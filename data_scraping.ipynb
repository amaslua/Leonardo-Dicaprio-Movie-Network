```python
import urllib.request
from bs4 import BeautifulSoup
import pandas as pd
```

```python
headers = {'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8', 'User-Agent': 'Mozilla/5.0'}

request = urllib.request.Request(url='https://www.themoviedb.org/person/6193-leonardo-dicaprio?language=en-US', headers=headers)
response = urllib.request.urlopen(request)
result = BeautifulSoup(response.read(), 'html.parser')
```

```python
movies = result.find('table', {'class': 'card credits'}).find_all('a', {'class':'tooltip'})
fullcast = []
for movie in movies:
    url = 'https://www.themoviedb.org' + movie['href']
    request = urllib.request.Request(url=url, headers=headers)
    response = urllib.request.urlopen(request)
    result = BeautifulSoup(response.read(), 'html.parser')

    x = result.find('p', {'class': 'new_button'})
    fullcast.append(x.findChild("a")['href'])
```

```python
data = []
for movie_url in fullcast:
    url = "https://www.themoviedb.org" + movie_url
    request = urllib.request.Request(url=url, headers=headers)
    response = urllib.request.urlopen(request)
    result = BeautifulSoup(response.read(), 'html.parser')
    find_movie = result.find('div', {'class': 'title ott_false'})
    if find_movie == None:
        find_movie = result.find('div', {'class': 'title ott_true'})
    movie_name = find_movie.getText(' ', strip=True)
    x = result.find('section', {'class': 'panel pad'}).find_all('a')
    for n in range(1, len(x), 2):
        actor_name = x[n].getText(' ')
        data.append([movie_name, actor_name])
```

```python
df = pd.DataFrame(data, columns = ["Movie", "Actor"])
selected_movies = ["The Wager", "Another Round", "Jim Jones", "Don't Look Up (2021)", "Clint Eastwood: A Cinematic Legacy (2021)", "Leonardo DiCaprio: Most Wanted! (2021)", "Teslafy Me (2019)", "QT8: The First Eight (2019)", "Ice on Fire (2019)", "And We Go Green (2019)", "Ryuichi Sakamoto: Coda (2017)", "Spielberg (2017)", "Before the Flood (2016)", "Koko: The Gorilla Who Talks to People (2016)", "The Revenant (2015)", "Scorsese's Goodfellas (2015)", "The Wolf of Wall Street (2013)", "The Great Gatsby (2013)", "Django Unchained (2012)", "Reflections on Titanic (2012)", "J. Edgar (2011)", "Inception (2010)", "Hubble (2010)", "Shutter Island (2010)", "Revolutionary Road (2008)", "Body of Lies (2008)", "Notes on an American Film Director at Work (2008)", "Everything's Cool (2007)", "Blood Diamond (2006)", "The Departed (2006)", "Earth to America (2005)", "A Concert for Hurricane Relief (2005)", "The Aviator (2004)", "Catch Me If You Can (2002)", "Gangs of New York (2002)", "Portrait of Leonardo: The Kid Who Took Hollywood (2000)", "Hangin' with Leo!! (2000)", "Romeo + Juliet (1996)", "Total Eclipse (1995)", "The Quick and the Dead (1995)", "This Boy's Life (1993)", "Mickey's Safety Club: Street Safe, Street Smart (1989)"]
df = df[df["Movie"].isin(selected_movies)]
df = df.reset_index(drop=True)
df
```
